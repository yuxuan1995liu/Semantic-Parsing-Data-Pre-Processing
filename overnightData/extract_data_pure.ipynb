```python
TRAIN_PATH1='basketball.paraphrases.train.examples'
TEST_PATH1='basketball.paraphrases.test.examples'
TRAIN_PATH2='blocks.paraphrases.train.examples'
TEST_PATH2='blocks.paraphrases.test.examples'
TRAIN_PATH3='calendar.paraphrases.train.examples'
TEST_PATH3='calendar.paraphrases.test.examples'
TRAIN_PATH4='publications.paraphrases.train.examples'
TEST_PATH4='publications.paraphrases.test.examples'
TRAIN_PATH5='housing.paraphrases.train.examples'
TEST_PATH5='housing.paraphrases.test.examples'
TRAIN_PATH6='recipes.paraphrases.train.examples'
TEST_PATH6='recipes.paraphrases.test.examples'
TRAIN_PATH7='restaurants.paraphrases.train.examples'
TEST_PATH7='restaurants.paraphrases.test.examples'
TRAIN_PATH8='socialnetwork.paraphrases.train.examples'
TEST_PATH8='socialnetwork.paraphrases.test.examples'
```

```python
with open(TRAIN_PATH1, 'r') as f:
    train_set_basketball = f.read().splitlines()
with open(TEST_PATH1, 'r') as f:
    test_set_basketball = f.read().splitlines()
with open(TRAIN_PATH2, 'r') as f:
    train_set_block = f.read().splitlines()
with open(TEST_PATH2, 'r') as f:
    test_set_block = f.read().splitlines()
with open(TRAIN_PATH3, 'r') as f:
    train_set_calendar = f.read().splitlines()
with open(TEST_PATH3, 'r') as f:
    test_set_calendar = f.read().splitlines()
with open(TRAIN_PATH4, 'r') as f:
    train_set_publications = f.read().splitlines()
with open(TEST_PATH4, 'r') as f:
    test_set_publications = f.read().splitlines()
with open(TRAIN_PATH5, 'r') as f:
    train_set_housing = f.read().splitlines()
with open(TEST_PATH5, 'r') as f:
    test_set_housing = f.read().splitlines()
with open(TRAIN_PATH6, 'r') as f:
    train_set_recipes = f.read().splitlines()
with open(TEST_PATH6, 'r') as f:
    test_set_recipes = f.read().splitlines()
with open(TRAIN_PATH7, 'r') as f:
    train_set_restaurants = f.read().splitlines()
with open(TEST_PATH7, 'r') as f:
    test_set_restaurants = f.read().splitlines()
with open(TRAIN_PATH8, 'r') as f:
    train_set_socialnetwork = f.read().splitlines()
with open(TEST_PATH8, 'r') as f:
    test_set_socialnetwork = f.read().splitlines()
```

```python
def remove_values_from_list(the_list, val):
    return [value for value in the_list if value != val]
#print(train_set)

def clear_data(train_set):
    clear_train_set = remove_values_from_list(train_set,'')
    clear_train_set1 = remove_values_from_list(clear_train_set,'  )')
    clear_train_set2 = remove_values_from_list(clear_train_set1,'(example')
    clear_train_set3 = remove_values_from_list(clear_train_set2,')')
    clear_train_set3 = remove_values_from_list(clear_train_set3,'  (targetFormula ')
    return clear_train_set3
```

```python
import copy 
def clean_data(clear_train_set3):
    list_copy=clear_train_set3.copy() 
    for i in range(len(clear_train_set3)): 
        if "utterance" in clear_train_set3[i]: 
            #list_copy[i] = "utterance "+clear_train_set3[i][14:-2]
            list_copy[i] = clear_train_set3[i][14:-2]
            #print (list_copy[i])
        elif "(original" in clear_train_set3[i]:
            #list_copy[i] = "original "+clear_train_set3[i][14:-2]
            list_copy[i] = clear_train_set3[i][13:-2]
            #print (list_copy[i])
        elif i%3==2:
            #list_copy[i] = "targetFormula "+clear_train_set3[i][4:]
            list_copy[i] = clear_train_set3[i][4:]
    return list_copy
        #print(list_copy[i])
#print(list_copy)
```

```python
train_basketball = clean_data(clear_data(train_set_basketball))
test_basketball = clean_data(clear_data(test_set_basketball))
train_block = clean_data(clear_data(train_set_block))
test_block = clean_data(clear_data(test_set_block))
train_calendar = clean_data(clear_data(train_set_calendar))
test_calendar = clean_data(clear_data(test_set_calendar))
train_publications = clean_data(clear_data(train_set_publications))
test_publications = clean_data(clear_data(test_set_publications))
train_housing = clean_data(clear_data(train_set_housing))
test_housing = clean_data(clear_data(test_set_housing))
train_recipes = clean_data(clear_data(train_set_recipes))
test_recipes = clean_data(clear_data(test_set_recipes))
train_restaurants = clean_data(clear_data(train_set_restaurants))
test_restaurants = clean_data(clear_data(test_set_restaurants))
train_socialnetwork = clean_data(clear_data(train_set_socialnetwork))
test_socialnetwork = clean_data(clear_data(test_set_socialnetwork))
```

```python
#a dictionary
def make_dict(list1):
    list_utterance = []
    list_original = []
    list_targetFormula = []
    for i in range(len(list1)):
        if i % 3 == 0:
            list_utterance.append(list1[i])
        if i % 3 ==1:
            list_original.append(list1[i])
        if i % 3 == 2:
            list_targetFormula.append(list1[i])
    dict = {}
    dict["utterance"]=list_utterance
    dict["original"] = list_original
    dict["targetFormula"] = list_targetFormula
    return dict
```

```python
final_dict_train={}
final_dict_test={}

final_dict_train["basketball"]=make_dict(train_basketball)
final_dict_train["block"]=make_dict(train_block)
final_dict_train["calendar"]=make_dict(train_calendar)
final_dict_train["publications"]=make_dict(train_publications)
final_dict_train["housing"]=make_dict(train_housing)
final_dict_train["recipes"]=make_dict(train_recipes)
final_dict_train["restaurants"]=make_dict(train_restaurants)
final_dict_train["socialnetwork"]=make_dict(train_socialnetwork)

final_dict_test["basketball"]=make_dict(test_basketball)
final_dict_test["block"]=make_dict(test_block)
final_dict_test["calendar"]=make_dict(test_calendar)
final_dict_test["publications"]=make_dict(test_publications)
final_dict_test["housing"]=make_dict(test_housing)
final_dict_test["recipes"]=make_dict(test_recipes)
final_dict_test["restaurants"]=make_dict(test_restaurants)
final_dict_test["socialnetwork"]=make_dict(test_socialnetwork)
```
