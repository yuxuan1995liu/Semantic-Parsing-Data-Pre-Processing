```python
me_sub = '.output_data_train.txt'
ref_sub ='./stem_atis/train.txt'
```

```python
with open(me_sub, 'r') as f:
    me_sub_train = f.read().splitlines()
    
with open(ref_sub, 'r') as f:
    ref_sub_train = f.read().splitlines()
```

```python
nl_me_sub = []
for strings in me_sub_train:
    k = strings.split(' \t')[0]
    k = k[1:]
    nl_me_sub.append(k)
print(nl_me_sub[:10])
```

```python
nl_ref_sub = []
for strings in ref_sub_train:
    j = strings.split('\t')[0]
    nl_ref_sub.append(j)
print(nl_ref_sub[:10])
```

```python
split_nl_me = []
for strings in nl_me_sub:
    k = strings.split(' ')
    split_nl_me.append(k)
#print(split_nl_me[:10])
split_me_clean = []
for index in range(len(split_nl_me)):
    k = []
    for index2 in range(len(split_nl_me[index])):
        if ('0' not in split_nl_me[index][index2]) and ('1' not in split_nl_me[index][index2]) and ('2' not in split_nl_me[index][index2]):
            k.append(split_nl_me[index][index2])
    split_me_clean.append(k)            
print(split_me_clean[:10])
```

```python
# create ref lib
split_nl_ref = []
for strings in nl_ref_sub:
    j = strings.split(' ')
    split_nl_ref.append(j)
split_ref_clean = []
for index in range(len(split_nl_ref)):
    k = []
    for index2 in range(len(split_nl_ref[index])):
        if ('0' not in split_nl_ref[index][index2]) and ('1' not in split_nl_ref[index][index2]) and ('2' not in split_nl_ref[index][index2]):
            k.append(split_nl_ref[index][index2])
    split_ref_clean.append(k)            

for index in range(len(split_ref_clean)):
    if split_ref_clean[index] == []:
        split_ref_clean[index] = ['a']
print(split_ref_clean[:10])
```

```python
words = 'abc'
wordss='abcd'
length = len([(t1, t2) for t1, t2 in zip(list(words),list(wordss)) if t1 == t2])
print(length)
```

```python
#split_me_clean - with untoken words
#split_ref_clean - with token words
relate_list = []
for index in range(len(split_ref_clean)):
    k = {}
    if split_ref_clean[index] == []:
        pass
    else:
        for index2 in range(len(split_ref_clean[index])):
            for index3 in range(len(split_me_clean[index])):
                length = len([(t1, t2) for t1, t2 in zip(split_ref_clean[index][index2],split_me_clean[index][index3]) if t1 == t2])
                if length > 2:
                    k[split_ref_clean[index][index2]] = split_me_clean[index][index3]
        relate_list.append(k)

for index in range(len(relate_list)):
    relate_list[index]['ani']='any'
    relate_list[index]['fli']='fly'
print(relate_list[909])
```

```python
k = 'one way'
k_split = k.split(' ')
k_split[0] = 'ones'
k = ' '.join(k_split)
print(k)
```

```python
string_sub = []
for index in range(len(nl_ref_sub)):
    k = nl_ref_sub[index].split(' ')
    #[airport],[one],[way]
    #['airport', 'one way', 'ci0 airport', 'ground ci0', 'code al0', 'ci0 airport', 'al0 fn0', 'describ ap0', 'ground transport ci0', 'ground transport ci0'
    for index2 in range(len(k)):
        if k[index2] in relate_list[index].keys():
            k[index2] = relate_list[index][k[index2]]
    ks = ' '.join(k)
    string_sub.append(ks)
print(string_sub)
```

```python
output =[]
for index in range(len(ref_sub_train)):
    j = string_sub[index] + '\t' + ref_sub_train[index].split('\t')[1]
    output.append(j)
print(output)
```

```python
# final output
with open('.output_try_train.txt', 'w',encoding='utf-8') as file:
    for i in output:
        file.write( i+ '\n')
```
