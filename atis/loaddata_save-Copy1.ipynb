```python
train_set_fl = []
for strings in train_set:
    train_set_fl.append(strings.split('\t')[1])
print(train_set_fl[:2])
```

```python
import re
#mke:_ap 
#denver:_ci
#boston:_ci )
token_list= []
for strings in train_set_fl:
    k = re.findall(r'[a-zA-Z0-9_]+\:\_[a-zA-Z0-9_]+',strings)
    #print(k)
    if k!=[]:
        token_list+=k
#append is to combine to lists. 
#if we have a = [1,2,3]
#b=[4,5,6]
#a+b = [1,2,3,4,5,6]
#a.append(b) = [1,2,3,[4,5,6]]
#but when you do a.append(b[0]) it becomes [1,2,3,4]
token_list =list(set(token_list))
for index in range(len(token_list)):
    token_list[index] = token_list[index].replace(' )','')
    token_list[index] = token_list[index].replace(' ','')
print(token_list)
```

```python
token = []
msg = []
for tokens in token_list:
    token.append(tokens.split(':_')[1])
token = list(set(token))
print(token)
```

```python
token_dict = {}
for tokens in token:
    token_language = []
    for strings in token_list:
        if strings[-2:] == tokens:
            token_language.append(strings[:-4])
            token_language = list(set(token_language))
        token_dict[tokens] = token_language

del(token_dict['mf'])
del(token_dict['dc'])
del(token_dict['pd'])
del(token_dict['hr'])
del(token_dict['cl'])
del(token_dict['i'])
del(token_dict['e'])
del(token_dict['t'])
del(token_dict['rc'])
token_dict['rc'] = ['s', 'b','ls','s/','ap/57','d/s','ap/80','sd/d','ap/68','ap/55','ap57','ap80','ap68','ap55']
for keys in token_dict.keys():
    for index in range(len(token_dict[keys])):
        token_dict[keys][index] = token_dict[keys][index].replace('_',' ')
#print(token_dict)
```
