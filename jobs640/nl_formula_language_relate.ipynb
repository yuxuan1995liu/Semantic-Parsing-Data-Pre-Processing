```python
#input reference data
with open(input_train_data,encoding='utf8') as f:
    train_data = f.read().splitlines()
with open(input_test_data,encoding='utf8') as f:
    test_data = f.read().splitlines()
with open(reference_data_train,encoding='utf8') as f:
    reference_train_data = f.read().splitlines()
with open(reference_data_test,encoding='utf8') as f:
    reference_test_data = f.read().splitlines()
```

```python
# Train
from nltk.stem import PorterStemmer as stemmer
# slice natural language from reference data
reference_train_nl = [line.split('\t')[0] for line in reference_train_data]

# remove punctuation from source train nl
import string
en_puns = string.punctuation
source_train_nl = []
for line in train_data:
    for p in en_puns:
        line = line.replace(p, ' ')
    line = ' '.join([t for t in line.split(' ') if len(t) > 0])
    source_train_nl.append(line)
def match(refer_nl):
    for p in en_puns:
        refer_nl = refer_nl.replace(p, ' ')
    tk_refer_nl = [t for t in refer_nl.split(' ') if len(t)>0]
    #print(tk_refer_nl)
    #tk_refer_nl is the original list as words
    stem_refer_nl = [stemmer().stem(t) for t in tk_refer_nl]
    #is the stem words of the original list
    #source_train_nl is the original hand-modified list
    match_lengths = {}
    for j in range(len(source_train_nl)):
        source_nl = source_train_nl[j]
        for p in en_puns:
            source_nl = source_nl.replace(p, ' ')
        tk_source_nl = [t for t in source_nl.split(' ') if len(t)>0]
        stem_source_nl = [stemmer().stem(t) for t in tk_source_nl]
#         print(stem_refer_nl)
#         print(stem_source_nl)
#         print('\n')
#stem_source_nl is the stemmer of source_train_nl

        if tk_refer_nl == tk_source_nl:
            return source_train_nl[j]
        elif stem_refer_nl == stem_source_nl:
            return source_train_nl[j]
    match_lengths = {}
        length = len([(t1, t2) for t1, t2 in zip(stem_refer_nl, stem_source_nl) if t1 == t2])
        match_lengths[length] = j
    max_len = max(list(match_lengths.keys()))
    target_index = match_lengths[max_len]
    return source_train_nl[target_index]
    
result_list = []
for i in range(len(reference_train_nl)):
    refer_nl = reference_train_nl[i]
    match_nl = match(refer_nl)
    if match_nl is not None:
        result_list.append(match_nl)
    else:
        print(refer_nl)
        break
    print('{}/{}'.format(i, len(reference_train_nl)))
```
